## Código de Transcrição, Tradução, ORFs e Extração de todas as proteínas

```python
def DNAtoRNA(dna):
    """
    Função que realiza a transcrição de
    uma sequência de DNA
    """
    DNA = dna.upper()
    RNA = DNA.replace("T", "U")
    return RNA

gencode = {
    'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M',
    'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
    'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K',
    'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',
    'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L',
    'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
    'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q',
    'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
    'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
    'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
    'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E',
    'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
    'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S',
    'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
    'TAC':'Y', 'TAT':'Y', 'TAA':'_', 'TAG':'_',
    'TGC':'C', 'TGT':'C', 'TGA':'_', 'TGG':'W'}

def DNAtoAMINO(dna):
    """
    Função que recebe uma sequência
    de DNA e devolve a sequência
    de aminoácidos correspondentes
    """
    dna2 = dna.upper()
    codons = []
    lista = []
    for i in range (0, len(dna2), 3):
        if len(dna2[i : i + 3]) == 3:
            codons.append(dna2[i : i + 3])
    for codon in codons:
        lista.append(gencode.get(codon))
    return ''.join(lista)

#ou então:
#def dna2amino(dna):
    #return ''.join([gencode.get(dna[p : p + 3],'') for p in range(0, len(dna), 3)])

def comp_inv(dna):
    """
    Função que recebe uma sequência
    de DNA e devolve o complementar inverso
    """
    return dna.lower().replace("a", "T").replace("c", "G").replace("g", "C").replace("t", "A")[::-1]

def get_orfs(dna):
    """
    Função que recebe uma sequência de DNA
    e devolve as 6 ORFs correspondentes
    """
    return [DNAtoAMINO(dna[p:]) for p in range(3)] + [DNAtoAMINO(comp_inv(dna)[p:]) for p in range(3)]

def get_orf_prot(orf):
    """
    Função que retorna as possíveis proteinas
    através das orfs fornecidas
    """
    is_prot = False
    prot=''
    prots=[]
    for A in orf:
        if is_prot:
            prot += A
            if A == '_':
                prots.append(prot)
                is_prot = False
                prot=''
        else:
            if A == 'M':
                is_prot = True
                prot = "M"
    return set(prots)

def get_prots(amino):
    """
    Função que recebe uma sequência
    de aminoácidos e retorna todas as
    proteínas possíveis associadas
    """
    proteins = []
    b = amino.upper()
    new_amino = b.split("_")
    for aa in new_amino:
        for idx, a in enumerate(aa):
            if a == "M":
                proteins.append(aa[idx:])
    return proteins

def get_all_prots(dna):
    res = set()
    for prots in [get_orf_prot(orf) for orf in get_orfs(dna)]:
        res |= prots
    return res
```

## Código para Alinhamentos Globais e Locais

```python
import io

class Mat:
    """
    Implementação de uma matriz
    """
    def __init__(self, rows, cols):
        self.mat = [[0 for c in range(cols)]
                    for r in range(rows)]
    def numRows (self): return len(self.mat)
    def numCols (self): return len(self.mat[0])
    def __str__(self):
        """
        Devolve a matriz como uma string
        """
        return '\n'.join(' '.join(str(val) for val in row)
                         for row in self.mat)
    def __repr__(self):
        """
        Utilizado no repl quando se pede para ver a matriz
        """
        return str(self)
    def __getitem__ (self, n):
        """
        Interface para a indexação []
        """
        return self.mat[n]
    
def subst(x, y):
    return 3 if x == y else -1
  
class NW:
    """
    Alinhamento global entre duas sequências
    """
    def __init__(self, s1, s2, g = -4):
        self.s1 = s1
        self.s2 = s2
        self.mat = Mat(len(s1) + 1, len(s2) + 1)
        self.tr  = Mat(len(s1) + 1, len(s2) + 1)
    
        for L in range(len(s1)):
            self.mat[L + 1][0] = g * (L + 1)
            self.tr[L + 1][0]  = 'C'

        for C in range(len(s2)):
            self.mat[0][C + 1] = g * (C + 1)
            self.tr[0][C + 1]  = 'E'

        for L, x1 in enumerate(s1):
            for C, x2 in enumerate(s2):
                possiveis = [
                    self.mat[L  ][C    ] + subst(x1, x2),   # Diagonal
                    self.mat[L+1][C    ] + g,               # Esquerda
                    self.mat[L  ][C + 1] + g,               # Cima
                ]
                dirs = "DEC"

                self.mat[L + 1][C + 1] = max(possiveis)
                self.tr[L + 1][C + 1] = dirs[possiveis.index(self.mat[L + 1][C + 1])]

    def rebuild(self):
        L = len(self.s1)
        C = len(self.s2)
        S1 = ""
        S2 = ""
    
        dirs = {
            'D' : (-1, -1),
            'E' : ( 0, -1),
            'C' : (-1,  0)
        }

        while L > 0 or C > 0:
            DL, DC = dirs[self.tr[L][C]]
        
            if self.tr[L][C] == "D":
                S1 = self.s1[L - 1] + S1
                S2 = self.s2[C - 1] + S2
            elif self.tr[L][C] == "E":
                S1 = '-' + S1
                S2 = self.s2[C - 1] + S2
            else:
                S1 = self.s1[L - 1] + S1
                S2 = '-' + S2        

            L += DL
            C += DC

        return S1, S2

    def __repr__(self):
        cols = "-" + self.s2
        lins = "-" + self.s1
        with io.StringIO("") as S:
            print(' ', *cols, sep = '   ', file = S)
            for L, linha in zip(lins, self.mat):
                print(L, *[f'{x:3d}' for x in linha], file = S)

            print(file = S)

            print(' ', *cols, file = S)
            for L, linha in zip(lins, self.tr):
                print(L, *linha, file = S)

            return S.getvalue()
        
## Código para Alinhamento Local

class ScoreParams:
    '''
    Função que permite definir o score para cada um dos parametros:
    gap
    match
    mismatch
    '''
    def __init__(self,gap,match,mismatch):
        self.gap = gap
        self.match = match
        self.mismatch = mismatch

def getMatrix(sizeX,sizeY):
    '''
    Função que cria uma matriz de zeros com tamanho proporcional
    '''
    matrix = []
    for i in range(len(sizeY)+1):
        subMatrix = []
        for j in range(len(sizeX)+1):
            subMatrix.append(0)
        matrix.append(subMatrix)
    return matrix

def localAlign(x,y,score):
    '''
    Função que permite preencher a nossa matriz com os scores e
    determinar o melhor com a melhor posição
    '''
    matrix = getMatrix(x,y)
    best = 0
    optLoc = (0,0)

    for i in range(1,len(y)+1):
        for j in range(1,len(x)+1):
            matrix[i][j] = max(
                matrix[i][j-1] + score.gap,
                matrix[i-1][j] + score.gap,
                matrix[i-1][j-1] + (score.match if x[j-1] == y[i-1] else score.mismatch),
                0
                )

            if matrix[i][j] >= best:
                best = matrix[i][j]
                optLoc = (i,j)

    return best, optLoc, matrix

def printMatrix(matrix):
    '''
    Função que imprime a matriz com
    os scores do alinhamento local
    '''
    for i in range(len(matrix)):
        print(matrix[i])
    print()
```

## Código BLAST

```python
def query_map(query, w):
    """
    Função que retorna um dicionário
    com subsequências como chaves de tamanho w
    e valores como os seus indíces
    """
    tam = len(query)
    res = {}
    for chave, offset in [(query[p: p + w], p) for p in range(0, tam - w + 1)]:
        if chave not in res: res[chave] = []
        res[chave].append(offset)
    return res

def get_all_offsets(s1, s2):
    """
    Função que retorna todos os offsets
    de s1 e s2
    """
    w = len(s1)
    res = []
    for p in range(0, len(s2) - w + 1):
        if s2[p : p + w] == s1:
            res.append(p)
    return res

def hits(qm, seq):
    """
    Função que retorna List[hits] onde
    hit = (o1, o2), sendo o1 o 
    índice na query e o2 o índice
    na sequência
    """
    res = []
    for chave, offsets in qm.items():
        for o_query in offsets:
            for o_seq in get_all_offsets(chave, seq):
                res.append((o_query, o_seq))
    return res

def extend_hit_dir(query, seq, o1, o2, direction):
    """
    Função que retorna (o1, o2, matches, count)
    """
    matches = 0
    count = 0
    while o1 >= 0 and o2 >= 0 and o1 < len(query) and o2 < len(seq):
        matches += 1 if query[o1] == seq[o2] else 0
        count += 1
        if 2 * matches < count:
            return o1, o2, matches, count
        o1 += direction
        o2 += direction
    return o1 - direction, o2 - direction, matches, count

def extend_hit(query, seq, hit, w):
    """
    Função que retorna (o1, o2, size, matches)
    onde o size é o tamanho total da expansão
    """
    o1, o2 = hit
    left  = extend_hit_dir(query, seq, o1 - 1, o2 - 1, -1)
    right = extend_hit_dir(query, seq, o1 + w, o2 + w, +1)

    O1, O2, ML, SL = left
    _,   _, MR, SR = right

    return O1, O2, w + SL + SR, ML + w + MR
```

## Código para Alinhamentos Múltiplos

```python
def consensus(s1,s2):
    """
    Função que calcula a sequência consenso entre
    duas seqs
    """
    res = ""
    for x, y in zip(s1,s2):
        if x == y:
            res += x
        elif x == "-":
            res += y
        else:
            res += x
    return res

def multiple_align(s1, s2, s3):
    """
    Função que apresenta o alinhamento múltiplo
    entre 3 sequências, utilizando a função
    consensus e a classe NW
    """
    a1, a2 = NW(s1, s2, g = -1).rebuild()
    n1 = consensus(a1,a2)
    n2, a3 = NW(n1, s3, g = -1).rebuild()
    x = NW(s1, n2, g=-1).rebuild()[0]
    y = NW(s2, n2, g=-1).rebuild()[0]
    z = NW(s3, n2, g=-1).rebuild()[0]
    return x, y, z

multiple_align(SEQ1[0:35], SEQ2[20:80], SEQ3[15:50])
```

## Código para Enzimas de Restrição

```python
enzimas = {"R":"[GA]", "Y":"[CT]", "M":"[AC]", "K":"[GT]", "S":"[GC]", "W": "[AT]", "B":"[^A]", "D":"[^C]", "H":"[^G]", "V":"[^T]", "N":"[ACGT]"}

enzima = input()

import re

def converte(enz):
    """
    Função que converte os caracteres sem ser "ACGT"
    nos nucleotidos correspondentes, se for o caso
    """
    return ''.join([enzimas.get(x,x) for x in enz])

def enz_corte(enzima):
    """
    Função que separa a enzima pelo local de corte
    e aplica a função converte definida em cima
    nas duas partes
    """
    e1, e2 = enzima.split('^')
    return f"(?<={converte(e1)}){converte(e2)}"

seq = [SEQ1, SEQ2, SEQ3]

for s in seq:
    pos = [n.span()[0] for n in re.finditer(enz_corte(enzima), s)]
    print(pos)
    print([s[i:j] for i,j in zip([0, *pos], [*pos, len(s)])])
```

## Motifs por Expressões regulares e por PROSITE

```python
def find_zync_finger(seq):
    """
    Função que procura um determinado padrão com expressões regulares
    e permite saber quantas vezes o padrão aparece em seq
    """
    from re import search
    regexp = "C.H.[LIVMFY]C.{2}C[LIVMYA]"
    mo = search(regexp , seq)
    if mo != None:
        return mo.span()[0]
    else: return "-1"
    
def test_zync_finger():
    """
    Função que testa a Função find_zync_finger
    """
    seq = "HKMMLASCKHLLCLKCIVKLG"
    print(find_zync_finger(seq))
```

```python
def find_prosite(seq, profile):
    """
    Função que assume como argumento o padrão Prosite, permitindo
    assim procurar e contar quantas vezes o determinado padrão aparece
    na seq
    """
    from re import search
    regexp = profile.replace("−","")
    regexp = regexp.replace("x",".")
    regexp = regexp.replace("(","{")
    regexp = regexp.replace(")","}")
    mo = search(regexp , seq)
    if mo != None: return mo.span()[0]
    else: return "-1"
    
def test_prosite():
    """
    Função que testa a Função find_prosite
    """
    seq = "HKMMLASCKHLLCLKCIVKLG"
    print(find_prosite(seq,"C−x−H−x−[LIVMFY]−C−x(2)−C−[LIVMYA]"))
```

```python
def contagens(seqs, chars, pseudo = 0):
    """
    Função que conta quantas vezes os diferentes caracteres
    aparecem nas dadas sequências
    """
    return [{x : s.count(x) + pseudo for x in chars} for s in zip(*seqs)]

def pwm(seqs, pseudo = 0):
    """
    Função que permite calcular as pseudocontagens dos caracteres das várias sequências
    """
    tam = len(seqs)
    chars = set(''.join(seqs))
    count = contagens(seqs, chars, pseudo)
    return [{k : v / sum(d.values()) for k, v in d.items()} for d in count]
```

```python
pwm(seqs, pseudo = 0.5)
```

```python
def pssm(seqs, pseudo = 0.5):
    """
    Função que calcula o logaritmo das pseudocontagens (pwm)
    """
    P = pwm(seqs, pseudo)
    nchars = len(set(''.join(seqs)))
    from math import log2
    return [{k: log2(v/(1/nchars)) for k, v in d.items()} for d in P]
```
